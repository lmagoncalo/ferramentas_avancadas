```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import  LabelEncoder

train_data = pd.read_csv("../input/ferramentas-avancadas-projeto/train_set.csv")
test_data = pd.read_csv("../input/ferramentas-avancadas-projeto/test_set.csv")

train_data = train_data.drop(['track_id', 'artists', 'album_name', 'track_name', "track_genre"], axis=1)
test_data = test_data.drop(['artists', 'album_name', 'track_name', "track_genre"], axis=1)

X_train = train_data.drop(['popularity'], axis=1)
y_train = train_data["popularity"]

X_test = test_data.drop(['popularity', 'track_id'], axis=1, errors='ignore')

model = LinearRegression()

model.fit(X_train, y_train)

prediction = model.predict(X_test)

test_data["popularity"] = prediction

test_data = test_data.drop(['mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'duration_ms', 'explicit', 'danceability', 'energy', 'key', 'loudness', 'valence', 'tempo'], axis=1)

test_data.to_csv('submission.csv', index=False)
```
