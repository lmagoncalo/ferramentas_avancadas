```python
from sklearn.linear_model import LinearRegression


X = np.array([[1], [2], [3], [4], [5]])  # X tem que ter sempre 2 dimensoes
print(X.shape)
y = np.array([2, 4, 6, 8, 10])  # y tem que ter o mesmo tamanho do X
print(y.shape)

model = LinearRegression()
model.fit(X, y)

print(model.coef_)
print(model.intercept_)
```

```python
X_new = np.array([[6], [7]])  # X tem que ter sempre 2 dimensoes
print(X_new.shape)

y_previsto = model.predict(X_new)
print(y_previsto)
```

```python
import matplotlib.pyplot as plt


plt.scatter(X, y, color="blue", label="Data")
plt.plot(X_new, y_previsto, color="red", label="Model Prediction")
plt.legend()
plt.show()
```

```python
model = LinearRegression()

X = np.array([[1], [2], [3], [4], [5]])
y = 3 * X + 1

model.fit(X, y)

print(model.coef_)
print(model.intercept_)
```
